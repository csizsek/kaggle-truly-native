```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn import cross_validation
from sklearn.metrics import roc_auc_score
import xgboost as xgb
from sklearn import grid_search
from sklearn import preprocessing
%pylab inline
```

```python
w = pd.read_csv('tmp/engineered_features.csv')
```

```python
f0 = pd.read_csv('tmp/0.csv')['file']
f1 = pd.read_csv('tmp/1.csv')['file']
f2 = pd.read_csv('tmp/2.csv')['file']
f3 = pd.read_csv('tmp/3.csv')['file']
f4 = pd.read_csv('tmp/4.csv')['file']
f5 = pd.read_csv('tmp/5.csv')['file']
```

```python
f03 = pd.concat([f0, f1, f2, f3])
f03.describe()
```

```python
df03 = DataFrame({'file': f03})
```

```python
d03 = pd.merge(df03, w, on='file', how='inner')
d03.info()
```

```python
df4 = DataFrame({'file': f4})
d4 = pd.merge(df4, w, on='file', how='inner')
d4.info()
```

```python
w = None
```

```python
trl = d03.sponsored
trl.index = d03.file
tr = d03
del tr['sponsored']
tr.index=d03.file
del tr['file']
```

```python
tr = tr.fillna(0)
tr = DataFrame(preprocessing.scale(tr))
tr.describe().T
```

```python
m = KNeighborsClassifier(n_neighbors=31, p=1, weights='distance')
```

```python
m.fit(tr, trl)
```

```python
te = d4
del d4['sponsored']
te.index = d4.file
del te['file']
te = te.fillna(0)
te = DataFrame(preprocessing.scale(te))
te.describe().T
```

```python
preds = m.predict_proba(te)
```

```python
sub = DataFrame({'sponsored': preds[:,1]})
sub.index = te.index
sub.info()
```

```python
sub.to_csv('output/count_okapilink_knn_pred_4.csv')
```
